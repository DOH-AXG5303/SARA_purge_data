```python
#import of all excel files into pandas (dictionary with key=date, value=df)

dataframe_dict = {}
failed_list = []

for name in os.listdir(general_path):
    
    name_search = re.search(r"\A(Sara-Alert-Purge-Eligible-Export-)([0-9]{4}-[0-9]{2}-[0-9]{2}).*(\.xlsx$)", name)
    
    if name_search:
        try:
            df = pd.read_excel(general_path + name, sheet_name = "Edit Histories", engine = "openpyxl")
            dataframe_dict[name_search.group(2)] = df
            print(name_search.group(2) + "SUCCESS")
        except:
            failed_list.append(name)
            print(name + "FAIL")     
    else:
        pass
```

```python
#which dataframes do not match column reference (jan24)

jan_24 = set(dataframe_dict["2022-01-24"].columns)

for key, value in dataframe_dict.items():
    clms_set = set(value.columns)
    
    if clms_set == jan_24:
        print(f"{key} -- Columns match")
    
    else:
        print(f"{key} does not match")
```

```python
#adding empty columns so that all dataframes match in prep for concat
may_19 = set(dataframe_dict["2021-05-19"].columns)
jan_24 = set(dataframe_dict["2022-01-24"].columns)

missing_clms = jan_24 - may_19

for key, value in dataframe_dict.items():
    df = value.copy()
    clms_set = set(value.columns)
    
    if clms_set == jan_24:
        #print(f"{key} -- Columns match")
        pass
    
    else:
        for i in missing_clms:
            df[i] = np.nan #add nans for columns names that are missing
        
        dataframe_dict[key] = df
        
```

```python
df_list = [value for key, value in dataframe_dict.items()]
final_df = pd.concat(df_list)
final_df = final_df.reset_index()
```

```python
final_df = final_df.drop(columns = ["index"])
```

```python
final_df = final_df.drop_duplicates()
```

```python
final_df = final_df[final_df["History Type"].isin(['Enrollment', 'Monitoring Change', 'Record Automatically Closed'])].copy()
```

```python
final_df = final_df.reset_index(drop = True)
```
